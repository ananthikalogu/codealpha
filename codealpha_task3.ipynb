```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = sns.load_dataset("titanic")

print("Sample Data:")
print(df.head())

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="class", hue="class", palette="Set2", legend=False)
plt.title("Passenger Count by Class")
plt.show()


plt.figure(figsize=(6,4))
sns.histplot(df['age'].dropna(), kde=True, bins=30, color="skyblue")
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(data=df, x="sex", y="survived", hue="sex", palette="pastel", legend=False)
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Probability")
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(data=df, x="class", y="survived", hue="class", palette="muted", legend=False)
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Probability")
plt.show()


plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
