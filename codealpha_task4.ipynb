```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("titanic")


print("Dataset Preview:")
print(df.head())
print("\nDataset Info:")
print(df.info())

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="class", palette="Set2", hue="class", legend=False)
plt.title("Passenger Count by Class")
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df['age'].dropna(), bins=30, kde=True, color="purple")
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(data=df, x="sex", y="survived", palette="pastel", hue="sex", legend=False)
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Probability")
plt.show()
plt.figure(figsize=(6,4))
sns.barplot(data=df, x="class", y="survived", palette="muted", hue="class", legend=False)
plt.title("Survival Rate by Class")
plt.ylabel("Survival Probability")
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
